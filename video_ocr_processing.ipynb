```python
# Import necessary libraries
import cv2
import easyocr
import psutil
import torch
import time
import numpy as np
import matplotlib.pyplot as plt
```

```python
from google.colab import files
# Upload the video file from your local system
uploaded = files.upload()

# Get the uploaded video file name
video_path = list(uploaded.keys())[0]
print(f"Video uploaded: {video_path}")
```

```python
# Initialize EasyOCR reader with GPU (if available)
reader = easyocr.Reader(['en'], gpu=True if torch.cuda.is_available() else False)
```

```python
# Capture video from the uploaded file
cap = cv2.VideoCapture(video_path)

# Get video properties
frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
fps = cap.get(cv2.CAP_PROP_FPS)
print(f"Total frames: {frame_count}, FPS: {fps}")
```

```python
# Performance metrics initialization
total_frames_processed = 0
frames_skipped = 0
total_ocr_time = 0
gpu_memory_usage = []
cpu_usage = []

frame_texts = []
```

```python
# Read only first 10 frames from the video and perform OCR
max_frames = 10  # Limit to first 10 frames
processed_frames = 0

while cap.isOpened() and processed_frames < max_frames:
    ret, frame = cap.read()
    if not ret:
        break

    processed_frames += 1

    start_time = time.time()

    # Perform OCR on the current frame
    results = reader.readtext(frame)
    frame_texts.append(results)

    # Time taken for OCR on this frame
    ocr_time = time.time() - start_time
    total_ocr_time += ocr_time

    # Extract the text from OCR results
    extracted_text = "\n".join([text[1] for text in results])

    # Print the extracted text for the frame
    print(f"Extracted Text for Frame {processed_frames}:")
    print(extracted_text)
    print("-" * 50)

    # Convert frame to RGB for displaying with Matplotlib
    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # Create a plot to display the image and the extracted text
    plt.figure(figsize=(10, 6))
    plt.imshow(frame_rgb)
    plt.title(f"Frame {processed_frames}")
    plt.axis('off')

    # Display extracted text on the image
    plt.figtext(0.1, 0.1, f"Extracted Text: {extracted_text}", wrap=True, fontsize=10, color="white", ha="left", va="top")

    plt.show()
```

```python
# Initialize EasyOCR reader with CPU (disable GPU)
reader = easyocr.Reader(['en'], gpu=False)  # Set gpu=False for CPU usage
```

```python
# Initialize performance tracking variables
total_frames_processed = 0
frames_skipped = 0
total_ocr_time = 0
frame_texts = []
```

```python
# Open the video file
cap = cv2.VideoCapture(video_path)

# Check if the video was loaded correctly
if not cap.isOpened():
    print("Error: Unable to open video file.")
else:
    # Get video properties
    frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    fps = cap.get(cv2.CAP_PROP_FPS)
    print(f"Total frames: {frame_count}, FPS: {fps}")

    # Initialize counters and storage
    total_frames_processed = 0
    frame_texts = []

    # Start performance tracking time
    start_time = time.time()

    # Process the video frames and perform OCR
    while cap.isOpened() and total_frames_processed < 10:
        ret, frame = cap.read()
        if not ret:
            break

        total_frames_processed += 1

        # Perform OCR on the current frame
        ocr_start_time = time.time()
        results = reader.readtext(frame)
        frame_texts.append(results)

        # Time taken for OCR on this frame
        ocr_time = time.time() - ocr_start_time

        # Convert frame to RGB for displaying with Matplotlib
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

        # Extracted text from OCR results
        extracted_text = "\n".join([text[1] for text in results])

        # Create a plot to display the image and the extracted text
        plt.figure(figsize=(10, 6))
        plt.imshow(frame_rgb)
        plt.title(f"Frame {total_frames_processed}")
        plt.axis('off')

        # Display the extracted text as well
        plt.figtext(0.1, 0.1, f"Extracted Text:\n{extracted_text}", wrap=True, fontsize=10, color="white", ha="left", va="top")

        # Show the frame with the extracted text
        plt.show()

        # Print the extracted text in the console
        print(f"Frame {total_frames_processed}: {extracted_text}")

    # Release the video capture object
    cap.release()
```
